In this example, we will re-use the logistic growth problem from before, but configure it to perform likelihood-free Bayesian inference. <br>
The process for instantiating the backend is identical to `ScipyBackend`. 
If the don't provide priors, the `pyABCBackend` will construct log-normal priors with a constant `sigma` around the initial values.

```python
from logistic_growth import prob
from mempyfit import pyABCBackend

abc = pyABCBackend(prob)
abc.plot_priors()
```

We can `max_total_nr_simulations` to be suitable to our inference problem. 

```python
abc.run(max_total_nr_simulations = 50_000)
```

```python
report = abc.report()
```
